```python
# Declaring Perceptron Classifier
class Perceptron:
    """
    Parameters: information which will be passed through constructor and other functions
    ------------------------
    eta: float
        Learning Rate (between 0.0 and 1.0)
    n_iter: int
        Number of Iteration Over Training Dataset, instead of loop until converge of original Perceptron Algorithm
    random_state: int
        Random Number Generator seed for random weight initialisation
    
    
    Attributes: variables for inter state check and information capture
    -------------------------
    w_: 1d-array
        weights after fitting
    b_: scalar
        bias after fittig
    errors_: list
        Number of misclassification updates, in each epoch
    
    """
    
    def __init__(self, eta=0.01, n_iter=50, random_state=1):
        self.eta = eta
        self.n_iter = n_iter
        self.random_state=random_state
    
    def fit(self,X,y):
        """
        Parameters:
        -----------
        X: {array-like}, shape=[n_example,n_features]
            Training Vector where n_examples is the number of examples,
            and n_features is number of feaures for each training example
            
        y: {array-like}, shape=[n_examples]
            Target Values for each of the training examples
        
        """
        rgen = np.random.RandomState(self.random_state)
        # Initiate Weight Vector with dimension equal to number of features
        self.w_ = rgen.normal(loc=0.0, scale=0.01, size=X.shape[1]) 
        self.b_ = np.float_(0.)
        self.errors_ = []
        
        for _ in range(self.n_iter):
            errors= 0
            for xi, target in zip(X,y):
                update = self.eta * (target - self.predict(xi))
                self.w_ += update * xi
                self.b_ += update
                errors += int(update!=0.0)
            self.errors_.append(errors)
        return self
    
    def net_input(self,X):
        """
        Calculate Net Input
        """
        return np.dot(X, self.w_) + self.b_
    
    def predict(self,X):
        "Return class label after unut step"
        return np.where(self.net_input(X) >=0.0, 1, 0)
```

```python
ppn = Perceptron(eta=0.01, n_iter=10)
ppn.fit(X,y)
plt.plot(range(1, len(ppn.errors_)+1), ppn.errors_, marker='o')
plt.xlabel('Number of Epochs')
plt.ylabel('Number of Misclassification / Updates')
plt.show()
```

```python
#Plot Decision Regions
from matplotlib.colors import ListedColormap
def plot_decision_regions(X,y,classifier,resolution=0.02):
    #Setup Marker Generator and Color Map
    markers = ('o','s','^','v','<')
    colors = ('red','blue','lightgreen','gray','cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])
    
    #Plot the decision surface
    x1_min, x1_max = X[:,0].min() - 1 , X[:,0].max() + 1
    x2_min, x2_max = X[:,1].min() - 1 , X[:,1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution), 
                          np.arange(x2_min, x2_max, resolution))
    lab = classifier.predict(np.array([xx1.ravel(),xx2.ravel()]).T)
    lab = lab.reshape(xx1.shape)
    plt.contourf(xx1,xx2, lab, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())
    
    #Plot Training Data
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y==cl, 0], y=X[y==cl,1],
                    alpha=0.8, c=colors[idx], marker=markers[idx],
                    label=f'Class {cl}: {iris.target_names[cl]}',
                   edgecolor='black')
```

```python
plot_decision_regions(X,y,ppn)
plt.xlabel('Sepal Length [cm]')
plt.ylabel('Petal Length [cm]')
plt.legend(loc="upper left")
plt.show()
```

## Adaline Implementation

We will now look at another __Single-Layer Neural Network__ known as __ADAptive LInear NEuron__ or __Adaline__ in short. This algorithm was published few year after Perceptron Algorithm, by Bernard Widrow and his doctoral student Tedd Hoff.

The algorithm is worth knowing, because it illustrates the key concept defining and minimising the continuous loss functions, and as such lays groundword for understanding other machine learning algorithms for classification, such as logistic regression, SVMs, multi layer perceptron (MLP), as well as linear regression models.

The main differences between Adaline Classifier and Perceptron classifier are:
* Weights are updated based on the Linear Activation Function, rather than a Unit Step Function. The Linear Activation function is simply the identity function of the net input.
* While the linear activation function is used for learning the weights, for making the final prediction the threshold function (Unit Step function) is still used

### Loss Function / Objective Function of Adaline Algorithm
The Adaline algorithm compares the true class labels with the linear activation function's continuous valued output to compute the model error and update the weights. In contrast, the perceptron compares the true class labels to the predicted class labels.

This brings us to one of the key ingredient of any machine learning algorithm, which is a defined __objective function__, that is to be optimized during the learning process. The main advantage of having continuous linear function as Activation Function, instead of Unit Step Function, is that the loss function becomes differentiable. Another important characteristic is that it is convex, and thus __gradient descent__ algorithm can be used to find the weights that minimize the loss function to classify the data


Although Adaline Algorithm is identical to Perceptron, we should note that
* the output of Activation Function = Identity of Net Input = Net Input, is a real number and not an integer class label.
* the weight update is calculated based on all all the examples in training set, instead of updating the parameters incrementally after each training example.


```python
# Implemention of Adaline Algorithm
class AdalineGD:
    """
    Adaptive Linear Neuron Gradient Descent
    
    Parameters
    -------------
    eta: float
        Learning Rate
        
    n_iter: int 
        number of passes over training dataset
        
    random_state: int
        Random number generator seed for random weight initialization.
    
    Attributes
    -------------
    w_: 1d-array
        Weights after fitting
    
    b_: scalar
        Bias unit after fitting
    
    losses_: list
        Mean squared error loss function values after each epoch
    """
    
    def __init__(self, eta=0.01, n_iter=50, random_state=42):
        self.eta = eta
        self.n_iter = n_iter
        self.random_state = random_state
    
    def fit(self, X, y):
        """
        Fit the Training Data
        
        Parameters:
        -----------
        X: {array-like}, shape=[n_example,n_features]
            Training Vector where n_examples is the number of examples,
            and n_features is number of feaures for each training example
            
        y: {array-like}, shape=[n_examples]
            Target Values for each of the training examples
        
        
        Return:
        ------
        self: object
        
        """
        rgen = np.random.RandomState(self.random_state)
        self.w_ = rgen.normal(loc=0.0, scale=0.01, size=X.shape[1])
        self.b_ = np.float_(0.)
        self.losses_ = []
        
        for i in range(self.n_iter):
            net_input = self.net_input(X)
            output = self.activation(net_input)
            errors = y - output
            self.w_ += self.eta * 2.0 * X.T.dot(errors) / X.shape[0]
            self.b_ += self.eta * 2.0 * errors.mean()
            loss = (errors ** 2).mean()
            self.losses_.append(loss)
        return self
    
    def net_input(self,X):
        """
        Calculate net input
        """
        return np.dot(X, self.w_) + self.b_
    
    def activation(self,X):
        return X
    
    def predict(self,X):
        "Return Class Label after Unit Step Function"
        return np.where(self.activation(self.net_input(X))>=0.5, 1, 0)
```

```python
#Plotting Loss against two different Learning Rates of Adaline Implementation
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,4))
ada1 = AdalineGD(n_iter=15, eta=0.1).fit(X,y)
ax[0].plot(range(1,len(ada1.losses_)+1), np.log10(ada1.losses_), marker='o')
ax[0].set_xlabel("Number of Epochs")
ax[0].set_ylabel("Log(Mean Squared Error)")
ax[0].set_title('Adaline - Learning Rate 0.1')
ada2 = AdalineGD(n_iter=15, eta=0.001).fit(X,y)
ax[1].plot(range(1,len(ada2.losses_)+1),ada2.losses_, marker='o')
ax[1].set_xlabel("Number of Epochs")
ax[1].set_ylabel("Mean Squared Error)")
ax[1].set_title('Adaline - Learning Rate 0.001')
plt.show()

           
```

Notice 
* Left Hand side: high learning rate resulting into increase of Training Loss (as we are overshooting the global minima)
* Right Hand side: lower learning rate will require of higher number of epochs for reaching global minima

```python
### SGD Adaline Implementation
class AdalineSGD:
    """
    Adaptive Linear Neuron with SGD
    
    Parameter:
    ------------
    eta: float
        Learning Rate
    
    n_iter: int
        Number of Iterations over the whole training dataset
    
    shuffle: bool
        Shuffles training data every epoch if True to prevent cycles
    
    random_state: int
        Random Number Generator seed for random weight initialization.
        
    Attributes:
    ------------
    w_: 1d-array
        Weights after training
    
    b_: Scalar
        Bias unit after training
    
    losses_: list
        Mean squared errorloss function value averaged over all training examples in 
        each epoch
    """
    
    def __init__(self, eta=0.01, n_iter=15, shuffle=True, random_state=None):
        self.eta=eta
        self.n_iter=n_iter
        self.shuffle=shuffle
        self.random_state= random_state
        self.w_initialized = False
    
    def fit(self, X, y):
        """
        Fit training data
        
        Parameters:
        --------------
        X: {array-like}, shape=[n_examples, n_features]
            Training vectors, where n_examples is the number of examples
            and n_features is the number of features
        y: array-like, shape=[n_examples]
            Target Values
        
        Return
        ------
        self: object
        """
        
        self._initialize_weights(X.shape[1])
        self.losses_ = []
        for i in range(self.n_iter):
            if self.shuffle:
                X,y = self._shuffle(X,y)
            losses = []
            for xi, target in zip(X,y):
                losses.append(self._update_weights(xi,target))
            avg_loss = np.mean(losses)
            self.losses_.append(avg_loss)
        return self
    
    def _shuffle(self,X,y):
        r = self.rgen.permutation(len(y))
        return (X[r], y[r])
    
    def _update_weights(self,xi,target):
        output = self.activation(self.net_input(xi))
        error = target - output
        self.w_ += self.eta * 2.0 * xi * (error)
        self.b_ += self.eta * 2.0 * (error)
        loss = error**2
        return loss
    
    def _initialize_weights(self,m):
        self.rgen = np.random.RandomState(self.random_state)
        self.w_ = self.rgen.normal(loc=0.0, scale=0.01, size=m)
        self.b_ = np.float_(0.)
        self.w_initalized=True
        
    def net_input(self,X):
        return np.dot(X,self.w_) + self.b_
    
    def activation(self,X):
        return X
    
    def predict(self,X):
        return np.where(self.activation(self.net_input(X))>=0.5, 1, 0)
    
    def partial_fit(self,X,y):
        """
        Fit Training data without reinitializing the weights
        """
        if not self.w_initialized:
            self._initialize_weights(X.shape[1])
            
        if y.ravel().shape[0]>1:
            for xi,target in zip(X,y):
                self._update_weights(X,y)
        else:
            self._update_weights(X,y)
        return self
```

```python
ada_sgd = AdalineSGD(n_iter=15, eta=0.01, random_state=1).fit(X_std,y)
    
```

```python
plot_decision_regions(X_std,y,classifier=ada_sgd)
plt.title('Adaline - Stochastic Gradient Descent')
plt.xlabel(f'Sepal Length - cm (standardized)')
plt.ylabel(f'Petal Length - cm (standardized)')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

```python
plt.plot(range(1, len(ada_sgd.losses_)+1), ada_sgd.losses_, marker='o')
plt.xlabel('Epochs')
plt.ylabel('Mean Square Loss')
plt.title('Training Loss for Epochs')
plt.show()
```

As you can see, average loss goes down pretty quickly, and the final decision boundary after 15 epochs looks similar to full-batch gradient descent Adaline. 

Additionally, the shuffle method which we have used generate the random permutation of indices, and are used to shuffle our feature matrix and training examples.

Also partial_fit method can be used in an online learning scenario with streaming data.

Also noted in terms of Gradient Descent approach you have following approaches (we will cover these details in later blogs):
* Gradient Descent => whole of the training example is processed in each epoch
* Stochastic Gradient Descent => one epoch consists of traversing each training example one by one, and making weight updates after every training example, usually multiple epochs are run
* Batch Gradient Descent => lies in between two approaches, instead of processing each training example one-by-one, mini-batches of training data is processed together
